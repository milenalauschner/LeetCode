Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.

Example 1:

Input: nums = [1,1,1,2,2,3], k = 2

Output: [1,2]

Example 2:

Input: nums = [1], k = 1

Output: [1]

Example 3:

Input: nums = [1,2,1,2,1,2,3,1,3,2], k = 2

Output: [1,2]

```python

nums1 = [1,2,1,2,1,2,3,1,3,2]
nums2 = [1,1,1,2,2,3]

def frequency(array, k):
    freq = {}
    seen = set()
    for number in array:
        freq[number] = freq.get(number, 0) +1
        seen.add(number)
    
    seen_list= list(seen)
   
    freq_k = []

    for idx in range(len(seen_list)):
        if freq[seen_list[idx]]>= k:
            freq_k.append(seen_list[idx])

    return freq_k

frequency(nums2,2)

## Complexity O(n)
```
