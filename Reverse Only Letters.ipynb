```python
#Two Pointer Solution = Sliding Window
#When pointers are in the same value it stops

def reverse_letters(s):
    s = list(s) #transformed in array
    print(s)
    left, right = 0, len(s) - 1
    #-1 because it is indexed at zero

    while left < right:
        if not s[left].isalpha():
            left += 1
        elif not s[right].isalpha():
            right -= 1
        else:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1

    return ''.join(s) #transform in a list again

lista = "a-bC-dEf-ghIj"

print(reverse_letters(lista))
```
