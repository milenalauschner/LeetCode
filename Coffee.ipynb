```python
lista1 = "aaaaabbbbbbdddeeeaaaccc"

def count(word):
    Dict = {}
    for letter in word:
        Dict[letter] = Dict.get(letter,0)+1

    result = ""

    for key,value in Dict.items():
        print(key)
        print(value)
        result = result + f"{key}{value}" 

    return result

count(lista1)
```
