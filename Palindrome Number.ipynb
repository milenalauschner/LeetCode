Given an integer x, return true if x is a palindrome, and false otherwise.

 

Example 1:

Input: x = 121
Output: true
Explanation: 121 reads as 121 from left to right and from right to left.
Example 2:

Input: x = -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.
Example 3:

Input: x = 10
Output: false
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.

```python
#if it is negative it will never be a palindrome

x = -121
def palindrome (number):
    if number <0:
        return False
    #NECESSARY to avoid ValueError: invalid literal for int() with base 10: '121-'
    if number == int(str(number)[::-1]):
        return True

palindrome(x) 

```

```python

#if it is negative it will never be a palindrome

x = "leonardo"
def palindrome (word):
    inverted_word = word[::-1]
    if word == inverted_word:
        return True
    else:
        return False

palindrome(x) 
```

sequence[start:stop:step]

start:
The index of the first element to include in the slice (inclusive). If omitted, it defaults to the beginning of the sequence (index 0).
stop:
The index of the first element to exclude from the slice (exclusive). If omitted, it defaults to the end of the sequence.
step:
The increment between elements in the slice. If omitted, it defaults to 1. A negative step value reverses the order of the slice.
