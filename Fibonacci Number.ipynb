The Fibonacci numbers, commonly denoted F(n) form a sequence, called the Fibonacci sequence, such that each number is the sum of the two preceding ones, starting from 0 and 1. That is,

F(0) = 0, F(1) = 1
F(n) = F(n - 1) + F(n - 2), for n > 1.
Given n, calculate F(n).

```python
def fibonacci (n):
    if n < 2:
        return n
    
    n = (n-1)+(n-2)
    
    if n > 30:
        return "Error"
    return n

fibonacci(56)
```

```python

def fib(n: int) -> int:
    if n < 2:
        return n
    a, b = 0, 1
    for num in range(n - 1):
        a, b = b, a + b #swapping values
        print("This is a", a)
        print("This is b", b)
    '''
    range(2, n+1) communicates the intent: “compute from i = 2 to i = n”. 
    If later you decide to use the index (e.g., for logging or debugging), 
    it’s already the right one. Using range(n-1) is fine but a bit less self-documenting.
    '''
    return b

fib(10)
```
