```python
rows = 3
cols = 4
nested_list_loop = []

for i in range(rows):
    inner_list = []
    for j in range(cols):
        inner_list.append(i * cols + j + 1) # Example: populate with sequential numbers
    nested_list_loop.append(inner_list)

print(nested_list_loop)
```
