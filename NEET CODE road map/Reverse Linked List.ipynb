## Two Pointers

Iterative

Time O(n)
Space O(1)


In Python, linked lists are not built-in types like lists or dictionaries â€” but you can implement them manually using classes.

```python
class Node:
    def __init__(self, data):
        self.data = data  # Value of the node
        self.next = None  # Pointer to the next node

class LinkedList:
    def __init__(self):
        self.head = None  # Start of the list

    def reverseList(head):
        prev, curr = None, head
        print("This is head:", head)

        while curr:
            temp = curr.next
            print("temporary key", temp)
            curr.next = prev
            prev = curr
            curr = temp
        return prev

    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=' -> ')
            current = current.next
        print('None')
```

Node represents each item in the list.

LinkedList manages the list as a whole.

next holds a reference to the next node, creating a chain.
