Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

Example 1:

Input: nums = [1, 2, 3, 3]

Output: true

Example 2:

Input: nums = [1, 2, 3, 4]

Output: false

```python
def contains_duplicate(array):
    seen = set()
    for number in array:
        if number not in seen:
            seen.add(number)
        else:
            print(seen)
            return True
    return False
    
array = [1, 2, 3, 3]

contains_duplicate(array)

```
