```python

def isPalindrome(s: str) -> bool:
    lower_s = s.lower()
    lista_s = list(lower_s)
    
    for idx in range(len(lista_s)-1):

        if idx < len(lista_s) and lista_s[idx] == ' ':
            lista_s.pop(idx)
        # idx < len(lista_s) prevent from accessing inexistent values in the array with len(lista_s)-1
    
    right, left = len(lista_s)-1, 0

    while right != left:
        if lista_s[right] == lista_s[left]:
            right -= 1
            left += 1
        else:
            return False
    return True


s = "Was it a car or a cat I saw"

isPalindrome(s)
```
