```python
array = [1, 3, 5, 2, 8, 1, 5]

def Three_Sum(arr):
    k = 3
    window_sum = sum(arr[:k])
    max_sum = window_sum
    for i in range(k, len(arr)):
        #print(f"Window sum before loop: {window_sum}")
        window_sum += arr[i] - arr[i - k] #arr[i - k] is 3 elements before arr[i]
        #print(f"this is k: {k}")
        #print(f"Window sum after loop: {window_sum} ({arr[i]} - {arr[i - k]})")
        max_sum = max(max_sum, window_sum)
    return max_sum

Three_Sum(array)
    
```
