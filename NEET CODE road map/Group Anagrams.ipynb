## Hash Table

Time complexity: 
O(m∗n)

Space complexity:
O(m) extra space.
O(m∗n) space for the output list.

```python
import collections


class Solution:
    def groupAnagrams(self, strs):
        res = collections.defaultdict(list)
        for s in strs:
            count = [0] * 26 #26 zeros = a.. z
            for c in s:
                count[ord(c) - ord('a')] += 1 #ord() gives the ASCII value of a character
                # a = 80 -> 0, 80-80
                # b = 81 -> 1, 81-80
            res[tuple(count)].append(s) #count is a list, so we change to tuples bcs tuples are hashable and can be used as keys of a dictionary
        return list(res.values())
    


strs = ["tops","stop","hat"]

result = Solution().groupAnagrams(strs)
print(result)
```

## Sorting

Time & Space Complexity
Time complexity: 
O(m∗nlogn)
Space complexity: 
O(m∗n)



```python
class Solution:
    def groupAnagrams(self, strs):
        res = defaultdict(list)
        for s in strs:
            sortedS = ''.join(sorted(s))
            res[sortedS].append(s)
        return list(res.values())
```
