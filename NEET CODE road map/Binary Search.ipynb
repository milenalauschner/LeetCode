You are given an array of distinct integers nums, sorted in ascending order, and an integer target.

Implement a function to search for target within nums. If it exists, then return its index, otherwise, return -1.

Your solution must run in O(logn) time.

Example 1:

Input: nums = [-1,0,2,4,6,8], target = 4

Output: 3
Example 2:

Input: nums = [-1,0,2,4,6,8], target = 3

Output: -1

## Recursive Binary Search

Time O(logn)
Space O(logn)

```python
# It is already SORTED

def binary_search(self,l,r, nums, target):
    if l > r:
        return -1
    
    m = l + (r - l) // 2
    
    if nums[m] == target:
        return m
    if nums[m] < target:
        return self.binary_search(m + 1, r, nums, target)
    return self.binary_search(l, m - 1, nums, target)

def search(self, nums, target: int) -> int:
    return self.binary_search(0, len(nums) - 1, nums, target)

```

## Iterative Binary Search

Time: O(logn)
Space: O(1)

```python
def search(nums, target: int) -> int:
    l, r = 0, len(nums) - 1
    print(f"left: {l}")
    print(f"right: {r}")
    while l <= r: #left can never be grater than right
        # (l + r) // 2 can lead to overflow -> the code never stops
        m = l + ((r - l) // 2) #This is mathematically equivalent to l+r//2
        print(f"Middle before: {m}")
        if nums[m] > target:
            r = m - 1
            print(f"new right: {r}")
        elif nums[m] < target:
            l = m + 1
            print(f"new left: {l}")
        else:
            return m
    return -1


nums = [-1,0,2,4,6,8]
target = 4

search(nums,target)
```

## Upper Bound

Time: O(logn)
Space: O(1)

```python
def search(nums, target: int) -> int:
        
        l, r = 0, len(nums)

        while l < r:
            m = l + ((r - l) // 2)
            if nums[m] > target:
                r = m
            elif nums[m] <= target:
                l = m + 1
        return l - 1 if (l and nums[l - 1] == target) else -1
```

## Lower Bound

Time: O(logn)
Space: O(1)

```python
def search(nums, target: int) -> int:

    l, r = 0, len(nums)

    while l < r:
        m = l + ((r - l) // 2)
        if nums[m] >= target:
            r = m
        elif nums[m] < target:
            l = m + 1
    return l if (l < len(nums) and nums[l] == target) else -1
```

## Built-In Function

Time: O(logn)
Space: O(1)

```python
import bisect

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        index = bisect.bisect_left(nums, target)
        return index if index < len(nums) and nums[index] == target else -1
```
