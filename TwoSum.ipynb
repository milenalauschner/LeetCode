With an array of 4 numbers [11,15,2,7] always return the 2 numbers that the sum is 9.

Solution with complement idea and 

```python
numbers = [11,15,2,7, -2,-6]

def find_pair_with_sum(numbers, target=9):
    seen = set()
    for number in numbers:
        complement = target - number
        seen.add(number)
    
        if complement in seen:
            return [complement, number]

    return None  # If no pair is found

# Example usage
result = find_pair_with_sum(numbers)
print(result)  
```

```python
from typing import List


nums = [11, 15, 2, 7, -2]
target = 9
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hasher = {}
        for idx, num in enumerate (nums) :
            if hasher.get(num) is not None: return [hasher.get(num), idx]
            hasher[target-num] = idx


Solution().twoSum(nums, target)
```
