## Bubble Sort

Bubble Sort has a time complexity of O(n²), which is not efficient for large arrays

```python
arr = [5, 3, 8, 1, 2]

# Bubble Sort
n = len(arr)
for i in range(n):
    print(f"this is i = {i}")
    for j in range(0, n - i - 1):
        # range(n - i - 1) also works
        print(f"this is j: {j}")
        if arr[j] > arr[j + 1]:
            # Swap
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
            print(arr)

print(arr)
```

## Insertion Sort

Still O(n²), but faster in practice for small/partially sorted arrays.

```python
arr = [5, 3, 8, 1, 2]

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        #or arr[j], arr[j + 1] = arr[j + 1], arr[j]
        j -= 1
    arr[j + 1] = key

print(arr)
```

## Merge Sort

Time Complexity: O(n log n)

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

arr = [5, 3, 8, 1, 2]
arr = merge_sort(arr)
print(arr)
```

## Quick Sort

Also faster O(n log n), though worst-case is O(n²) if not handled carefully.

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        left = [] # [x for x in arr[1:] if x <= pivot] 
        right = [] # [x for x in arr[1:] if x > pivot]

        for x in arr[1:]:
            if x <= pivot:
                left.append(x)  # x is each individual element from the sublist arr[1:]
        
        for x in arr[1:]:
            if x > pivot:
                right.append(x)
                
        return quick_sort(left) + [pivot] + quick_sort(right) #recursive function
        

# Example usage:
arr = [11, 2, 9, 7, 5, 6]
sorted_arr = quick_sort(arr)
print(sorted_arr)
```

Choose a pivot (in this case, the first element).

Partition the array:

left: elements less than or equal to the pivot.

right: elements greater than the pivot.

Recursively apply Quick Sort to left and right.

Combine results: quick_sort(left) + [pivot] + quick_sort(right)
